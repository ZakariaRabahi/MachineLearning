## Type of Problem

The problem we are addressing is predicting gokart rental checkouts based on various features such as date, hour, temperature, wind, humidity, etc. This is a supervised learning problem because we have labeled data (i.e., we know the number of rentals for each set of features).

Furthermore, this is a regression problem because the target variable (number of gokart rentals) is continuous rather than categorical. We aim to predict the actual rental count, not classify it into discrete categories.


```python
# Cell 1: Import Libraries
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


# Suppress FutureWarnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

# Cell 2: Load and Preprocess Data
We load the data from the CSV file, convert the `Date` column to datetime format, and extract day, month, and year from the date. Additionally, we replace infinite values with NaNs and drop them to ensure the dataset is clean for further analysis.


```python
# Cell 2: Load and Preprocess Data
# Load the data
data = pd.read_csv('train.csv')

# Convert Date to datetime
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

# Extract day, month, and year
data['Day'] = data['Date'].dt.day
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

# Replace infinite values with NaN and then drop them
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.dropna(inplace=True)
```

# Exploratory Data Analysis With Boxplot

In this section, we will perform exploratory data analysis using box plots. Box plots provide insights into the distribution of numerical features in the dataset and help identify outliers. This technique is essential for understanding the variability and central tendency of gokart rental counts in relation to various weather conditions.

To generate multiple box plots for all numerical features in the dataset, we will use the following numerical features: 'Temperature(C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)', 'Dew point temperature(C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)', and 'Gokart rental count'.

The box plots visualize the distribution of various numerical features in the dataset. Each plot highlights the median, quartiles, and potential outliers for a specific feature. The median is represented by the middle line in each box, which corresponds to the 50th percentile of the feature. The interquartile range (IQR) is represented by the box, containing the middle 50% of the data (from the 25th percentile to the 75th percentile). The whiskers extend from the box to show the range of the data, excluding outliers, typically representing 1.5 times the IQR. Individual points beyond the whiskers indicate outliers, which are values that deviate significantly from the rest of the data.



```python
# Cell 8: Box Plots for Numerical Features
import matplotlib.pyplot as plt
import seaborn as sns

# List of numerical features
numerical_features = ['Temperature(C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)',
                      'Dew point temperature(C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 
                      'Snowfall (cm)', 'Gokart rental count']

# Generate multiple box plots for all numerical features in the dataset
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features):
    plt.subplot(3, 3, i+1)
    sns.boxplot(x=data[feature])
    plt.title(f'{feature} Boxplot')
plt.tight_layout()
plt.show()

```
